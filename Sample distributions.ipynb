### Plot terminal distribution of a stock following GBM

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters for the GBM process
mu = 0.1   # Drift
sigma = 0.2 # Volatility

# Time settings
T = 1.0     # Time to maturity
N = 10000   # Number of samples
```

```python
# Generate GBM samples
np.random.seed(42)  # For reproducibility
S0 = 100  # Initial stock price
W = np.random.randn(N) * np.sqrt(T)
S = S0 * np.exp((mu - 0.5 * sigma**2) * T + sigma * W)

# Calculate the histogram and plot the PDF
plt.hist(S, bins=100, density=True, alpha=0.6, color='b')
plt.xlabel('Spot at T')
plt.ylabel('Probability Density')
plt.title('PDF of Geometric Brownian Motion')
plt.grid(True)
plt.show()
```
